```python
'''
    Example model from chapter 16 in Ben Lamberts Book
    A Student's Guide to Bayesian Statistics
'''
import pystan as stan
import arviz as az
import numpy as np
import matplotlib as plt
import scipy

N=100
Y=np.random.normal(1.5,0.2,N) #Data

#Stan model
model = stan.StanModel(file='Transformed.stan')
fit = model.sampling(data={'Y':Y,'N':N}, iter=400,chains=4)
print(fit)
fit.plot()

inf_data = az.convert_to_inference_data(fit)
az.plot_energy(inf_data)
az.plot_density(inf_data)
```
