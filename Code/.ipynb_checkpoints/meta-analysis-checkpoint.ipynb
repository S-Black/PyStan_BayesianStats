```python
'''
Problem 17.1 A meta-analysis of beta blocker trials
Ben Lambert 'A Student's Guide to Bayesian Statistics'
'''

import pystan as stan
import pandas as pd
import seaborn as sns
import numpy as np

#nice plot aesthetic
sns.set()  
%matplotlib inline

#Read data
data=pd.read_csv("C:\\all_data\\hierarchical_betaBlocker.csv")
data.head() # Preview the first 5 lines of the loaded data 
data.describe()

data_stan = {
    'N': data['N'][0],
    'Rt': data['rt'],
    'Rc': data['rc'],
    'Nt': data['nt'],
    'Nc': data['nc']
}

#Fit Stan heterogeneous model
model = stan.StanModel(file='meta.stan')
fit_het = model.sampling(data=data_stan,iter=1000,chains=4, seed=1)

#Fit Stan hierarchical model
model = stan.StanModel(file='meta_hierarchical.stan')
fit_hir = model.sampling(data=data_stan,iter=1000,chains=4, seed=1,control=dict(adapt_delta=0.95))

#Results
print(fit_het)
sns.violinplot(data=fit_het['delta'])     #distribution by trial plot
sns.distplot(np.mean(fit_het['delta'],0)) #histogram of average delta values

print(fit_hir)
sns.violinplot(data=fit_hir['delta'])     #distribution by trial plot
sns.distplot(np.mean(fit_hir['delta'],0)) #histogram of average delta values
```
