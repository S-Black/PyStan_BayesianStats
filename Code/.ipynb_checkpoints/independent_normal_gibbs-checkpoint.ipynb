```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pystan


data_path = "data/"
out_path = "output/"

#np.random.seed(101)

n_samples=5000 
n_var=5 
X=np.c_[np.ones((n_samples,1)),np.random.normal(7,1,(n_samples,n_var-1))] 
 
B = 8*np.random.uniform(0,1,n_var) #draws 5 random numbers from normal distribution and multiplies by 8 
Sigma = np.array([1.9]) 
E = np.sqrt(Sigma)*(np.random.normal(0,1,n_samples)) 
Y=X.dot(B.T)+(E) 


print("precision h",1/(E.dot(E)/n_samples))

# Hyperparameters
b0 = np.zeros((n_var))
V0 = np.ones((n_var))*100
nu0 = 4
s0 = 4
s02 = s0^2

        

data_in={ 
    'N':n_samples, 
    'K':n_var, 
    'X':X, 
    'y':Y,
    'beta_0':b0,
    'V_0':V0,
    'nu_0':nu0,
    's_02':s02
    } 



# Specify the model
modelspec = """
data { 
    int<lower=0> N;
    int<lower=0> K; 
    matrix[N, K] X; 
    vector[N] y;
    vector[K]  beta_0; // prior mean beta
    vector[K]  V_0; // prior covariance beta
    real  nu_0; // prior df h
    real  s_02; // inverse of prior mean h
    } 

transformed data{
  real alpha;  # hyperparameters to be used in gamma dist
  real beta_h;

  alpha = 0.5 * nu_0;
  beta_h = alpha * s_02;
}

parameters {
    vector[K] beta;
    real<lower=0> h;
}
transformed parameters {
    real<lower=0> sigma;
    sigma=1/h;
}

model{
  beta ~ normal(beta_0, V_0);
  h ~ gamma(alpha, beta_h);
  //y ~ normal(X * beta, 1/h);
  y ~ normal(X * beta, sigma);
}
"""


# Compile the model
sm = pystan.StanModel(model_code=modelspec)

# Estimate the model
fit = sm.sampling(data=data_in, iter=500, chains=4,verbose=True)



print(fit) 
fit.plot() 
```
