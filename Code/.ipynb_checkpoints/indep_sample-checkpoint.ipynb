```python
import pystan as stan
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#nice plot aesthetic
sns.set()  
%matplotlib inline

#Stan model
model = stan.StanModel(file='indep_samples.stan')
fit = model.sampling(data={'mu':10,'kappa': 5}, algorithm='Fixed_param',iter=4000,chains=1)
print(fit)
fit.plot()

data = fit['Y'] #extract sampled data

sns.distplot(data) #display histogram
```
